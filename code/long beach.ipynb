## Module 1: Loading the dataset and understanding its structure
In this module, we load the census style dataset into a pandas DataFrame and explore its basic structure. The goal is to understand what the data looks like before doing any analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

pd.set_option("display.max_columns", 8)
pd.set_option("display.width", 120)

df = pd.read_csv("long beach.csv")
```

```python
print("=== Head of the DataFrame ===")
print(df.head())
```

```python
print("\n=== Tail of the DataFrame ===")
print(df.tail())
```

```python
print("\n=== Shape of the DataFrame (rows, columns) ===")
print(df.shape)
```

```python
print("\n=== Column names ===")
print(df.columns)
```

```python
print("\n=== Data types of each column ===")
print(df.dtypes)
```

```python
print("\n=== DataFrame info summary ===")
df.info()
```

## Module 2: Basic summaries and missing values
In this module, we calculate simple summary statistics for numeric columns, look at the distribution of categorical columns, and investigate missing values.

```python
print("\n=== Basic numeric summary (describe) ===")
print(df.describe())
```

```python
print("\n=== Summary for categorical columns ===")
print(df.describe(include=["object"]))
```

```python
print("\n=== Value counts for 'location' ===")
print(df["location"].value_counts())
```

```python
print("\n=== Number of missing values per column ===")
print(df.isna().sum())
```

```python
print("\n=== Proportion of missing values per column ===")
print(df.isna().mean().round(3))
```

```python
df_property_nonmissing = df.dropna(subset=["property_value"])
print("\n=== Shape after dropping rows with missing 'property_value' ===")
print(df_property_nonmissing.shape)
```

```python
print("\n=== Proportion of missing values per column ===")
print(df_property_nonmissing.isna().mean().round(3))
```

```python
df_na = df.dropna()
print("\n=== Shape after dropping rows with missing ===")
print(df_na.shape)
```

```python
print("\n=== Proportion of missing values per column ===")
print(df_na.isna().mean().round(3))
```

## Module 3: Selecting rows and columns, filtering with conditions
We practice selecting specific columns and rows using column names and index positions. We also filter the DataFrame using logical conditions.

```python
ages = df["age"]
print("\n=== First 5 values of age column ===")
print(ages.head())
```

```python
cols_of_interest = ["age", "sex", "race_eth", "location", "income_to_poverty_ratio"]
small_df = df[cols_of_interest]
print("\n=== Sample of selected columns ===")
print(small_df.head())
```

```python
print("\n=== Using loc to select rows 0 to 4 and a subset of columns ===")
print(df.loc[0:4, ["age", "sex", "race_eth"]])
```

```python
print("\n=== Using iloc to select rows 0 to 4 and first 5 columns ===")
print(df.iloc[0:5, 0:10])
```

```python
senior_mask = df["age"] >= 65
seniors = df[senior_mask]
print("\n=== Seniors (age >= 65), first 5 rows ===")
print(seniors.head())
```

```python
senior_poor_mask = (df["age"] >= 65) & (df["tenure"] == "Owned")
senior_poor = df[senior_poor_mask]
print(senior_poor[["age", "below_poverty_line", "location"]].head())
```

## Module 4: Groupby and aggregation
We use groupby to compute aggregated statistics across groups such as race, location, or education level.

```python
print("\n=== Mean income_to_poverty_ratio by race_eth ===")
print(df.groupby("race_eth")["income_to_poverty_ratio"].mean().round(1))

print("\n=== Mean income_to_poverty_ratio by location ===")
print(df.groupby("location")["income_to_poverty_ratio"].mean().round(1))

print("\n=== Aggregated property_value by location (mean, median, count) ===")
print(df.groupby("location")["property_value"].agg(["mean", "median", "count"]).round(1))
```

## Module 5: Basic visualizations with pandas and matplotlib
We create simple visualizations directly from the DataFrame such as histograms, bar plots, and scatter plots.

```python
plt.figure(figsize=(8,4))
df["age"].hist(bins=30)
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```
